```python
def pyomo1():
  global model
  model=ConcreteModel()
  model.x=Var(range(5),initialize=2)
  model.x[0].setub(5)
  model.x[0].setlb(-5)
  model.x[2].setlb(0)
  model.x[2].setub(3)
  def obj(model):
    return (model.x[0]-1)**2 + (model.x[0]-model.x[1])**2 + (model.x[1]-model.x[2])**2 + (model.x[2]-model.x[3])**4 + (model.x[3]-model.x[4])**4
  model.objective=Objective(rule=obj,sense=minimize)
  model.constr1=Constraint(expr=model.x[0]+model.x[1]**2+model.x[2]**3 - 2 -3*(2**0.5)==0)
  model.constr2=Constraint(expr=model.x[1]-model.x[2]**2+model.x[3]+2-2*(2**0.5)==0)
  model.constr3=Constraint(expr=model.x[0]*model.x[4]-2==0)
```

```python
def result1():
  solver = SolverFactory('ipopt', executable='/content/ipopt')
  global results
  results = solver.solve(model)
  print(results)
  print('\n objective function value = ', model.objective())
  print('\n optimal values of the optimization variables')
  for i in range(5):
    print('x',i+1,'  =   ',model.x[i].value)
  print('\n constraints')
  print(model.constr1.display())
  print(model.constr2.display())
  print(model.constr3.display())
```

```python
def pyomo(p):
  global model
  model=ConcreteModel()
  model.x=Var(range(5),initialize=2**p)
  model.x[0].setub(5)
  model.x[0].setlb(-5)
  model.x[2].setlb(0)
  model.x[2].setub(3)
  def obj(model):
    return (model.x[0]-1)**2 + (model.x[0]-model.x[1])**2 + (model.x[1]-model.x[2])**2 + (model.x[2]-model.x[3])**4 + (model.x[3]-model.x[4])**4
  model.objective=Objective(rule=obj,sense=minimize)
  model.constr1=Constraint(expr=model.x[0]+model.x[1]**2+model.x[2]**3 - 2 -3*(2**0.5)==0)
  model.constr2=Constraint(expr=model.x[1]-model.x[2]**2+model.x[3]+2-2*(2**0.5)==0)
  model.constr3=Constraint(expr=model.x[0]*model.x[4]-2==0)
```

```python
def result():
  solver = SolverFactory('ipopt', executable='/content/ipopt')
  global results
  results = solver.solve(model)
  print(results)
  print('\n objective function value = ', model.objective())
  print('\n optimal values of the optimization variables')
  for i in range(5):
    print('x',i+1,'  =   ',model.x[i].value)
  print('\n constraints')
  print(model.constr1.display())
  print(model.constr2.display())
  print(model.constr3.display())
```
